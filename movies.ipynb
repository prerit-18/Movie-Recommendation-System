Remove stopwords and punctuations

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

re- regular expression

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]','',text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop= True)
```

```python
indices = pd.Series(df.index, index = df['title']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
tfidf_matrix=tfidf.fit_transform(df['tags'])
# print(tfidf_matrix.toarray())
```

#CosineSimilarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title, n=10):
    # Recommend n movies

    if title not in indices:
        return ['Movie not Found']

    idx = indices[title]

    sim_score = cosine_similarity(
        tfidf_matrix[idx],
        tfidf_matrix
    ).flatten()

    similar_idx = sim_score.argsort()[::-1][1:n+1]
    #Sort on the basis of indices
    return df['title'].iloc[similar_idx]
```

sim_score = cosine_similarity(

).flatten()

👉 A 1D array of similarity scores

👉 Between selected movie and all movies

sim_score.argsort()

argsort() returns the indices that would sort the array.

[::-1]
Reverses the array

[1:n+1]

This slicing does two things:

🔹 Why skip [0]?

Because:

Index 0 is the movie itself.

Similarity with itself = 1.0

We don’t want to recommend the same movie.

So we skip first element.
